```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def scrape_and_store_(filenamelist,category):
  driver_ =  webdriver.Chrome('chromedriver',options=options)
  driver_.implicitly_wait(15)
  wait = WebDriverWait(driver_, 15)
  for filename in filenamelist:
    try:
      driver_.get(root3 + filename + '.cha')
      wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "utterance")))
      elements = driver_.find_elements_by_class_name("utterance")
      #newfile = open("/content/drive/My Drive/stage2022textdata/"+ category + "/" + filename + ".txt", "w")
      newfile = open("/content/drive/My Drive/stage2022textdata/"+ category + "/" + filename[filename.rfind("/")+1:len(filename)]+ ".txt", "w")
      for element in elements: 
        newfile.write(element.text + "\n")
    except TimeoutError:
      return None
    newfile.close()
```

```python
## don't touch
scrape_and_store_(eigsti,"eigsti")
```
